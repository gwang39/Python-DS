## $\Delta$ 2.1 Run our first Python program in Jupyter

Hit `Run` or **"Shift + Enter"** to run the program (execute a cell)

> <font color='#D61A46'> Note: Don't worry too much about the coding part. That's what we will learn in the data science course.</font>

```python
import numpy as np
import argparse
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import random

%matplotlib inline


def loadDataSet(fileName):      #load function to parse tab-delimited floats
    dataMat = []                #empty data matrix, assume last column is target value
    f = open(fileName)
    for line in f.readlines():
        curLine = line.strip().split('\t')
        fltLine = list(map(float,curLine)) #map all elements to float()
        dataMat.append(fltLine)
    return dataMat


def kmeans_model(inputMatrix, init, n_clusters, n_init, max_iter):
    k_means = KMeans(init=init, n_clusters=n_clusters, n_init=ninit, max_iter=maxiter)
    k_means.fit(inputMatrix)
    print(k_means.cluster_centers_)
    return k_means


def plot_kmeans(k_means, n_clusters):
    fig = plt.figure(figsize=(8, 8))
    colors = ["#%06x" % random.randint(0, 0xFFFFFF) for x in range(n_clusters)]
    ax = fig.add_subplot(1,1,1)
    for k, col in zip(range(n_clusters), colors):
        clustmemship = k_means.labels_ == k
        cluster_center = k_means.cluster_centers_[k]
        ax.plot(datMat[clustmemship, 0], datMat[clustmemship, 1], 'w',
                markerfacecolor=col, marker='.', markersize=12)
        ax.plot(cluster_center[0], cluster_center[1], 'o', markerfacecolor=col,
                markeredgecolor='k', markersize=12)
    ax.set_title('KMeans')
    plt.show()


# Set kmeans paramters
numClusters = 3
data = './kmeans_random.txt'
init = 'k-means++'
maxiter = 100
ninit = 5

# Load data
datMat = np.mat(loadDataSet(data))

# Train kmeans
k_means = kmeans_model(datMat, init, numClusters, ninit, maxiter)

# Plot results
plot_kmeans(k_means, numClusters)
```

## $\Delta$ 2.4 Jupyter Magic

Lines and cells that starts with "%" are not interpreted like normal Python. They are MAGIC!!! 
> To Jupyter users: Magics are specific to and provided by the **IPython kernel**. Whether magics are available on a kernel is a decision that is made by the kernel developer on a per-kernel basis. To work properly, Magics must use a syntax element which is not valid in the underlying language. For example, the IPython kernel uses the % syntax element for magics as % is not a valid unary operator in Python.

> <font color='#D61A46'> Note: % Magic doesn't work in a regular python IDE. </font>

### <font color="#20czd6"> Magic: Going visual inline with Matplotlib

> Matplotlib cell magic let us tell Jupyter we want to see charts inline with the notebook. We will go details in the Python Day 2 courses.

```python
'''Plot 3D Contour Plot Inline'''
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib import cm

%matplotlib inline

fig = plt.figure(figsize=(9,9))
ax = fig.gca(projection='3d')
X, Y, Z = axes3d.get_test_data(0.05)
ax.plot_surface(X, Y, Z, rstride=8, cstride=8, alpha=0.3)
cset = ax.contourf(X, Y, Z, zdir='z', offset=-100, cmap=cm.coolwarm)
cset = ax.contourf(X, Y, Z, zdir='x', offset=-40, cmap=cm.coolwarm)
cset = ax.contourf(X, Y, Z, zdir='y', offset=40, cmap=cm.coolwarm)

ax.set_xlabel('X')
ax.set_xlim(-40, 40)
ax.set_ylabel('Y')
ax.set_ylim(-40, 40)
ax.set_zlabel('Z')
ax.set_zlim(-100, 100)

plt.show()
```

### What is `saspy`? 

> The `SASPy` module provides Python APIs to the SAS system. You can start a `SAS session` and `run analytics from Python` through a combination of object-oriented methods and Python magics.

Typically, programming with this module is performed with the Python functions that are provided by the package. For example, to view the first few rows of a data set, you can use the `head()` method. 

- However, if you are an experienced SAS programmer, there might be occasions in which you prefer to run SAS statements–such as running `PROC PRINT` to view the first few rows of a data set.

### Reference

> **`saspy`** documentation [[link]](https://sassoftware.github.io/saspy/getting-started.html#)
