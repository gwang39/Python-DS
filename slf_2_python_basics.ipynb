## $\Delta$ 4.3 Naming Conventions


### Naming Conventions
> 
**Classes** 
* (other than Python’s built-in classes) should have a name that serves as a **`singular noun`**, and should be **capitalized** (e.g., Date rather than date or Dates). When multiple words are concatenated to form a class name, they should follow the so-called **`“CamelCase” convention`** in which the first letter of each word is capitalized (e.g., **`CreditCard`**).       

> 
**Functions**  
* Functions including member functions of a class, should be **`lowercase`**.
* If multiple words are combined, they should be separated by **`underscores (e.g., make_payment)`**. The name of a function should typically be a verb that describes its affect. However, if the only purpose of the function is to return a value, the function name may be a noun that describes the value (e.g., sqrt rather than calculate sqrt).        

>  
**Paramters, Instance Variables, Local Varaibles**
* Names that identify an individual object (e.g., a parameter, instance variable, or local variable) should be a lowercase noun (e.g., price).
* Occasionally, we stray from this rule when using a single uppercase letter to designate the name of a data structures (such as tree T).  


>  
**Constants**
* Identifiers that represent a value considered to be a constant are traditionally identified using **`all capital letters`** and with underscores to separate words (e.g., **`MAX_SIZE`**)     

>   
**Encapsulation**
* Recall from our discussion of encapsulation that identifiers in any context that begin with a **`single leading underscore` (e.g., `_secret`)** are intended to suggest that they are only for “internal” use to a class or module, and not part of a public interface.

```python
%%SAS
data mydata.both(drop=rc);
  declare Hash Plan (); /* declare the name Plan for hash */
  rc = plan.DefineKey ('Plan_id'); /* identify fields to use as keys */
  rc = plan.DefineData ('Plan_desc'); /* identify fields to use as data */
  rc = plan.DefineDone (); /* complete hash table definition */
  do until (eof1) ; /* loop to read records from Plan */
    set mydata.plan end = eof1;
    rc = plan.add (); /* add each record to the hash table */
  end;
  
  do until (eof2) ; /* loop to read records from Members */
    set mydata.member end = eof2;
    call missing(Plan_desc); /* initialize the variable we intend to fill */
    rc = plan.find (); /* lookup each plan_id in hash Plan */
    output; /* write record to Both */
  end;
  stop;
run; 

proc print data=mydata.both;
run;
```

# <a name="cast"></a><font color='#347B98'> 7. Type Conversion <font size='3'>[Back to TOC](#toc-2)</font>

    

The process of converting the value of one data type (integer, string, float, etc.) to another data type is called **type conversion**. Python has two types of type conversion.
* Implicit Type Conversion
* Explicit Type Conversion

> `Implicit Type Conversion` is automatically performed by the Python interpreter.
> - Python avoids the loss of data in Implicit Type Conversion.    

> `Explicit Type Conversion` is also called Type Casting, the data types of object are converted using predefined function by user.
> - In Type Casting loss of data may occur as we enforce the object to specific data type.
> - We can convert between different data types by using different type conversion functions like `int(), float(), str()` etc.

## $\Delta$ 6.1 Implicit Type Conversion

```python
num_int = 123
num_flo = 1.23

print(num_int, 'is type', type(num_int))
print(num_flo, 'is type', type(num_flo))
```

```python
num_new = num_int + num_flo

print(num_new, 'is type', type(num_new))
```

## $\Delta$ 6.2 Explicit Type Conversion (Type Casting)

```python
num_int = 123
num_str = "456"

print("Data type of num_int:",type(num_int))
print("Data type of num_str:",type(num_str))

print(num_int+num_str)
```

In the above program,

We add two variable num_int and num_str.
As we can see from the output, we got typeerror. Python is not able use Implicit Conversion in such condition.
However Python has the solution for this type of situation which is know as Explicit Conversion.

```python
num_int = 123
num_str = "456"

print("Data type of num_int:",type(num_int))
print("Data type of num_str before Type Casting:",type(num_str))

num_str = int(num_str)
print("Data type of num_str after Type Casting:",type(num_str))

num_sum = num_int + num_str

print("Sum of num_int and num_str:",num_sum)
print("Data type of the sum:",type(num_sum))
```

## $\Delta$ 9.1 Objects and Names

`Name` (also called `identifier`) is simply a name given to objects. 
> **`Everything in Python is an object`**. 
> Name is a way to access the underlying object.

For example, when we do the assignment `a = 2`, here `2 is an object stored in memory` and a is the name we associate it with. We can get the address (in RAM) of some object through the built-in function, `id()`

To simply put it, **`namespace`**` is a collection of names`.
* In Python, you can imagine a namespace as a mapping of every name, you have defined, to corresponding objects.
* Different namespaces can co-exist at a given time but are completely isolated.

  * These different namespaces are isolated. Hence, the same name that may exist in different modules do not collide.

## $\Delta$ 10.3 `while loop`

The while loop in Python is used to iterate over a block of code `as long as the test expression (condition) is true`.

We generally use this loop when we don't know beforehand, the number of times to iterate.

<img src='https://cdn.programiz.com/sites/tutorial2program/files/whileLoopFlowchart.jpg'>

### <font color='#FC7307'> Example 1 </font>

```python
# Program to add natural
# numbers upto 
# sum = 1+2+3+...+n
```

```python
n = 10

# initialize sum and counter
sum = 0
i = 1

while i <= n:
    sum = sum + i
    i = i+1    # update counter

# print the sum
print("The sum is", sum)
```

### <font color='#FC7307'> Example 2 </font>

> if the number of participants does not exceed 10, keep adding new participant

```python
participants = ['shaohua','jack','alina','rob']
min_participants = 10
```

```python
num_of_participants = len(participants)

while num_of_participants < min_participants:
    participants.append('placeholder')
    num_of_participants = len(participants)
```
